# Fast AES Implementation: A High-Throughput Bitsliced Approach

## Abstract

- With the proposed bitsliced approach, each parallelization unit processes an unprecedented number of thirty-two 128-bit input data.
- the **ShiftRows** stage can be implicitly handled through input rearrangement and is simplified to the point where its computing process can be **neglected**.
- In addition, the need for look-up table based I/O operations, which are used by the Substitute Bytes stage is eliminated through using **S-box logic circuit**.
- We develop high-throughput CTR and ECB AES encryption/decryption on 6 CUDA-enabled GPUs, which achieve 1.47 and 1.38 Tbps of encryption throughput on Tesla V100 GPU, respectively.


## Introduction

In our proposed implementation model, we introduce a data representation scheme and use it in the alteration of the AES stages so that they have far **less costly Byte-wise** operations.
This is done by changing the representation of the data from the commonly used row-major to the **column-major representation** to achieve our predefined goal of high-throughput encryption.
