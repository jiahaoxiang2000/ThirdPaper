# motivation

the implementation care about three domains: security, cost and performance. First, we show the relationship between the three domains on the cipher implementation.

## relationship S-C-P

we use the sage math to draw the not accurate 3D figure to show the relationship between the three domains. 

The security is the X axis, the performance is the Y axis, and the cost is the Z axis. on the similar security level, the performance is better, the cost is higher. on the similar performance level, the security is better, the cost is higher. on the similar cost level, the security is better, the performance is better. 

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create mesh grid
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)
Z = 1/2*X + 1/2*Y

# Create mask for regions
ang30 = np.tan(np.pi/6)
ang60 = np.tan(np.pi/3)
mask1 = Y <= X*ang30
mask2 = (Y > X*ang30) & (Y <= X*ang60)
mask3 = Y > X*ang60

# Create figure
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


# Plot surfaces for each region
ax.plot_surface(X*mask1, Y*mask1, Z*mask1, color='blue', alpha=0.6)
ax.plot_surface(X*mask2, Y*mask2, Z*mask2, color='red', alpha=0.6)
ax.plot_surface(X*mask3, Y*mask3, Z*mask3, color='green', alpha=0.6)

# Customize appearance with better label visibility
ax.set_xlabel('Security', labelpad=15)
ax.set_ylabel('Performance', labelpad=15)
ax.set_zlabel('Cost', labelpad=15, rotation=90)

# Adjust label properties
ax.xaxis.label.set_size(12)
ax.yaxis.label.set_size(12)
ax.zaxis.label.set_size(12)

# Customize viewing angle
ax.view_init(elev=15, azim=220)  # Changed angles for better visibility
ax.grid(False)

# Optional: Set axis limits to focus view
# ax.set_xlim(0, 5)
# ax.set_ylim(0, 5)
# ax.set_zlim(0, 5)
plt.show()
```
